```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

Para saber se há variaveis redundantes usaremos correlação de pierson com um threshold de 0,85 tudo que for isso pra cima será considerado redundante por ter uma altissima correlação entre si

```python
target = "SeriousDlqin2yrs"
X = df.drop(columns=[target])

threshold = 0.85
corr_matrix = X.corr().abs()

# Pega os pares de variáveis com correlação alta
high_corr_pairs = np.where(corr_matrix > threshold)
high_corr_pairs = [(corr_matrix.columns[x], corr_matrix.columns[y])
                   for x, y in zip(*high_corr_pairs)
                   if x != y and x < y]  

# Extrai todas as variáveis envolvidas
redundantes = set([x for pair in high_corr_pairs for x in pair])

# Visualizar o heatmap
plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlação de Pearson entre variáveis")
plt.show()
print(" Pares altamente correlacionados:")

for a, b in high_corr_pairs:
    print(f"{a} - {b} (r={corr_matrix.loc[a,b]:.2f})")

print("\n Colunas possivelmente redundantes:", redundantes)
```

Para saber se alguma variavel é irelevante usaremos o metodo ANOVA F-Test onde os valores P < 0,05 serão considerado irrelevantes

```python
from sklearn.feature_selection import f_classif

# Preencher valores ausentes (ANOVA não aceita NaN)
X_filled = X.fillna(0)

# Calcular F e p-valores
f_values, p_values = f_classif(X_filled, df[target])

# Organizar resultados em tabela
anova_df = pd.DataFrame({
    "Feature": X.columns,
    "F_value": f_values,
    "p_value": p_values
}).sort_values(by="p_value")

print("Resultados da ANOVA F-Test:")
print(anova_df)

# Filtrar colunas irrelevantes (p > 0.05)
irrelevantes = anova_df[anova_df["p_value"] > 0.05]["Feature"].tolist()
print("\nColunas estatisticamente irrelevantes (p > 0.05):", irrelevantes)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Selecionar apenas colunas numéricas
numeric_cols = df.select_dtypes(include=['number']).columns

# Plotar boxplots
plt.figure(figsize=(12, 8))
for i, col in enumerate(numeric_cols, 1):
    plt.subplot(3, 4, i)  # ajusta número de linhas/colunas conforme o nº de variáveis
    sns.boxplot(y=df[col], color="skyblue")
    plt.title(col, fontsize=10)
    plt.tight_layout()

plt.suptitle("Boxplots - Distribuição e Outliers", fontsize=14, weight="bold", y=1.02)
plt.show()
```

| Variável                                                                        |                                                                                                  |
| ------------------------------------------------------------------------------- | --------------------------------------------------------------------------------------------------------------------------------- |
| **RevolvingUtilizationOfUnsecuredLines_log**                                    | A maioria das pessoas usa pouco crédito (caixa perto de 0), mas algumas têm uso altíssimo — outliers de uso de cartão.            |
| **DebtRatio_log**                                                               | A maior parte das pessoas tem dívida pequena, mas há casos com dívida centenas de vezes maior que a renda — outliers financeiros. |
| **MonthlyIncome_log**                                                           | A maioria tem renda normal (~R$ 3.000 a R$ 10.000), mas há alguns casos com rendas altíssimas (outliers positivos).               |
| **NumberOfTimes90DaysLate_log**                                                 | Quase todo mundo nunca atrasou (0), mas há quem tenha atrasado dezenas de vezes — outliers reais, não erros.                      |
| **NumberOfTime60-89DaysPastDueNotWorse_log** e **30-59DaysPastDueNotWorse_log** | A maioria das pessoas não atrasa, mas poucas têm histórico pesado de atrasos — esses pontos são os *casos críticos*.              |


```python
# Criar colunas transformadas (mantendo as originais)
for col in ['RevolvingUtilizationOfUnsecuredLines', 'DebtRatio', 'MonthlyIncome']:
    df[f'{col}_log'] = np.log1p(df[col])

# Visualizar boxplots das versões transformadas
plt.figure(figsize=(10, 4))

for i, col in enumerate(['RevolvingUtilizationOfUnsecuredLines_log', 
                         'DebtRatio_log', 
                         'MonthlyIncome_log'], 1):
    plt.subplot(1, 3, i)
    sns.boxplot(y=df[col], color='skyblue')
    plt.title(col, fontsize=10)

plt.suptitle("Boxplots após transformação logarítmica", fontsize=14, weight='bold')
plt.tight_layout()
plt.show()
```

```python
# Aplicar winsorização nas 3 variáveis mais críticas
cols = ['DebtRatio_log', 'RevolvingUtilizationOfUnsecuredLines_log', 'MonthlyIncome_log']

for col in cols:
    lower = df[col].quantile(0.01)
    upper = df[col].quantile(0.99)
    df[col] = df[col].clip(lower, upper)

# Ver novo boxplot pra mostrar resultado
plt.figure(figsize=(10,4))
for i, col in enumerate(cols, 1):
    plt.subplot(1, 3, i)
    sns.boxplot(y=df[col], color='lightgreen')
    plt.title(f"{col} (após winsorização)", fontsize=9)
plt.tight_layout()
plt.show()
```

```python
# Limitar no máximo a 20 atrasos (acima disso é ruído ou erro)
df['Pagamentos_Atrasados'] = np.where(df['Pagamentos_Atrasados'] > 20, 20, df['Pagamentos_Atrasados'])
```

```python
plt.figure(figsize=(3,4))
sns.boxplot(y=df['Pagamentos_Atrasados'], color='lightblue')
plt.title('Pagamentos_Atrasados (limitado)')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

```python
from scipy.stats import f_oneway, pearsonr

# ANOVA (variável categórica x contínua)
f_val, p_val = f_oneway(df['Pagamentos_Atrasados'], df['SeriousDlqin2yrs'])
print(f"ANOVA: F = {f_val:.3f}, p = {p_val:.5f}")

# Correlação de Pearson
corr, _ = pearsonr(df['Pagamentos_Atrasados'], df['SeriousDlqin2yrs'])
print(f"Correlação com alvo: {corr:.3f}")
```

```python
# 1️⃣ Corrigir casos de renda zero (substituir por NaN)
df.loc[df['MonthlyIncome'] == 0, 'MonthlyIncome'] = np.nan

# 2️⃣ Tratar casos de renda ausente com valor mediano
median_income = df['MonthlyIncome'].median()
df['MonthlyIncome'].fillna(median_income, inplace=True)

# 3️⃣ Corrigir DebtRatio — limitar a 1.0 (100% da renda)
df['DebtRatio_limited'] = np.where(df['DebtRatio'] > 1, 1, df['DebtRatio'])

# 4️⃣ Aplicar transformação logarítmica no DebtRatio ajustado
df['DebtRatio_log_win2'] = np.log1p(df['DebtRatio_limited'])

# 6️⃣ Comparar boxplots antes x depois da correção
plt.figure(figsize=(8,4))

plt.subplot(1,2,1)
sns.boxplot(y=df['DebtRatio_log'], color='salmon')
plt.title("DebtRatio_log_win (antes)", fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.subplot(1,2,2)
sns.boxplot(y=df['DebtRatio_log_win2'], color='lightgreen')
plt.title("DebtRatio_log_win2 (após limite e log)", fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.suptitle("Correção do DebtRatio — Opção 1 (Lógica + Log)", fontsize=13, weight='bold')
plt.tight_layout()
plt.show()
```

```python
for col in ['DebtRatio_log_win2', 
            'RevolvingUtilizationOfUnsecuredLines', 
            'MonthlyIncome']:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower) | (df[col] > upper)]
    print(f"{col}: {len(outliers)} outliers restantes ({len(outliers)/len(df)*100:.2f}%)")
```

```python
for col in ['DebtRatio_log', 'DebtRatio_log_win2']:
    print(col, df[col].mean(), df[col].median())
```

```python
#  Substituir o DebtRatio pela versão corrigida e tratada
df['DebtRatio'] = df['DebtRatio_log_win2']  # versão final limpa e log-transformada

#  Substituir as demais variáveis pelo padrão log + winsorizado
df['RevolvingUtilizationOfUnsecuredLines'] = df['RevolvingUtilizationOfUnsecuredLines_log']
df['MonthlyIncome'] = df['MonthlyIncome_log']

# Definir colunas principais (oficiais)
cols_principais = [
    'SeriousDlqin2yrs',
    'RevolvingUtilizationOfUnsecuredLines',
    'age',
    'Pagamentos_Atrasados',
    'DebtRatio',
    'MonthlyIncome',
    'NumberOfOpenCreditLinesAndLoans',
    'NumberRealEstateLoansOrLines',
    'NumberOfDependents'
]

print("\n Variáveis oficiais prontas para modelagem:\n")
print(df[cols_principais].head())

# 5️⃣ Remover colunas auxiliares antigas (logs, winsor, etc.)
df.drop(columns=[
    # versões intermediárias
    'DebtRatio_log', 'DebtRatio_log_win', 'DebtRatio_log_win2',
    'RevolvingUtilizationOfUnsecuredLines_log', 'RevolvingUtilizationOfUnsecuredLines_log_win',
    'MonthlyIncome_log', 'MonthlyIncome_log_win',
    'NumberOfTimes90DaysLate', 'NumberOfTimes90DaysLate_log',
    'NumberOfTime60-89DaysPastDueNotWorse', 'NumberOfTime60-89DaysPastDueNotWorse_log',
    'NumberOfTime30-59DaysPastDueNotWorse', 'NumberOfTime30-59DaysPastDueNotWorse_log'
], inplace=True, errors='ignore')

print("\n Dataset final consolidado e pronto para normalização/modelagem!")

# 6️⃣ (Opcional) Reordenar colunas para consistência
df = df[cols_principais]
```
